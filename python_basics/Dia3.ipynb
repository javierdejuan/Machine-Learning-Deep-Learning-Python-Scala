## aplicaiones de funciones sobre dataframes

```python
df_numerico = pd.DataFrame({"a":np.random.normal(0,1, (7,)), # (7,) crea un array de 7 elementos
                            "b":np.random.normal(1,3, (7,)),
                            "c":np.random.normal(2,5, (7,)),
                            "d":np.random.normal(-4,1,(7,)),
                            "e":np.random.normal(2,1, (7,))})
                            
```

```python
#sacar el numero mas grande de cada columna
```

```python
df_numerico.apply(lambda x: x.max())
```

```python
#sacar el numero mas grande de cada fila
```

```python
df_numerico.apply(lambda x: x.max(),axis=1) #axis =1 filas, axis = 0 columnas, por defecto
```

```python
df_numerico.applymap(lambda x: "numero mayor que 3" if x > 3 else "numero menor que 3") #applymap -> x son las celdas
```

```python
df_numerico.max()
```

```python
df_numerico.max(axis=1)
```

```python
df_numerico.std()
```

```python
df_numerico.std(axis=1)
```

```python
df_numerico.describe()
```

```python
import pandas_profiling
pandas_profiling.ProfileReport(df_numerico)
```

```python
df_numerico.corr(method="spearman")
```

```python
df_numerico.cov()
```

# pivoter y reformar dataframes

```python
mi_primer_dataframe
```

```python
df_dos=pd.DataFrame({"nombre":np.array(["julio","jose","luis","daniel","enano"]),
                     "altura":np.array([175.1,189,156,201.3,114]),
                    "casado/a":np.array([False,False,True,False,True])})
```

```python
df_dos
```

```python
# joins
mi_primer_dataframe.merge(df_dos,how="inner",on="nombre")
```

```python
mi_primer_dataframe.merge(df_dos,how="left",on="nombre")
```

```python
mi_primer_dataframe.merge(df_dos,how="right",on="nombre")
```

```python
df_joineado=mi_primer_dataframe.merge(df_dos,how="outer",on="nombre")
```
